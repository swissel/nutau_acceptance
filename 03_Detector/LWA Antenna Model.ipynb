```python
import nec_util
import matplotlib.pyplot as pyp
import os
import numpy as np
%matplotlib inline
print os.environ['LWA_MODELS']
```

```python
pastel_colors = ['#FF6666','#FFCC66','#CCFF66','#66FF66','#66FFCC','#66FFFF','#66CCFF','#6666FF','#CC66FF','#FF66FF','#FF6FCF']
bright_colors = ['#FF0000','#FF8000','#FFFF00','#80FF00','#00FF00','#00FF80','#00FFFF','#0080FF','#0000FF','#8000FF','#FF00FF']
bright_colors2 = bright_colors = ['#FF0000','#FF8000','#80FF00','#00FF80','#0080FF','#0000FF','#8000FF','#FF00FF','#FF0080']
pyp.rcParams['font.size']=16
pyp.rcParams['legend.labelspacing'] = 0.1
```

```python
nec_lwa_3m_gnd_z = nec_util.NECImpedance(os.environ['LWA_MODELS']+"LWA_DUAL_0-90DEG-DRIVE_300x300x1cmGRID_Symbols")
nec_lwa_3m_gnd_pat = nec_util.NECPattern(os.environ['LWA_MODELS']+"LWA_DUAL_0-90DEG-DRIVE_300x300x1cmGRID_Symbols.nec", 50)
```

```python
pyp.figure()
pyp.grid()
pyp.semilogx(nec_lwa_3m_gnd_z.freqs[1:], nec_lwa_3m_gnd_z.z.real[1:],'b', label="Real(Z)")
pyp.plot(nec_lwa_3m_gnd_z.freqs[1:], nec_lwa_3m_gnd_z.z.imag[1:],'r', label="Imag(Z)")
pyp.plot(nec_lwa_3m_gnd_z.freqs[1:], abs(nec_lwa_3m_gnd_z.z[1:]),'k', label="|Z|")
pyp.xlabel("Frequency (MHz)")
pyp.xlim(0,200)
pyp.ylabel('Impedance ($\Omega$)')
pyp.legend(loc='upper left')
pyp.title(r'Impedance for LWA Antenna 3x3m Ground Plane', fontsize=12)
```
